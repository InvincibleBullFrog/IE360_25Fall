```python
import cvxpy as cp
import numpy as np

Q = np.array([[2.0, 0.0],
              [0.0, 1.0]])      # use numpy arrays
q = np.array([-1.0, -2.0])

x = cp.Variable(2)
prob = cp.Problem(cp.Minimize(0.5*cp.quad_form(x, Q) + q @ x), [x >= 0])
prob.solve(solver=cp.MOSEK)      # or prob.solve() to let CVXPY choose

print(prob.status, prob.value, x.value, prob.solver_stats.solver_name)
```
