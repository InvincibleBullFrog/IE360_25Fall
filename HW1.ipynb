```python
import cvxpy as cp
import numpy as np

B = cp.Variable(nonneg=True, name="Illini Blizzards")
S = cp.Variable(nonneg=True, name="Caramel Sundaes")

constraints = [
    2*B + S <= 100,     # milk
    B + 1.5*S <= 150,   # sugar
    B >= S              # pride
]

objective = cp.Maximize(4*B + 2*S)
prob = cp.Problem(objective, constraints)

prob.solve(solver=cp.MOSEK)

print("status:", prob.status)
print("Illini Blizzards = ", round(float(B.value), 2))
print("Caramel Sundaes =", round(float(S.value), 2))
print("Revenue = $", round(float(prob.value), 2))
print("Milk used =", round(float(2*B.value + S.value), 2))
print("Sugar used =", round(float(B.value + 1.5*S.value), 2))
```
